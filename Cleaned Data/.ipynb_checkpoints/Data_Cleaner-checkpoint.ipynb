```python
def format_main_objective_as_list(objectives_string):
    if isinstance(objectives_string, str) and objectives_string.strip():
        objectives_list = [obj.strip() for obj in objectives_string.split(',')]
        formatted_list = ""
        for i, objective in enumerate(objectives_list):
            if objective: 
                formatted_list += f"{i + 1}. {objective}\n"
        return formatted_list.strip() 
    else:
        return "" 
```

```python
Chdf['Main Objective'] = Chdf['Main Objective'].apply(format_main_objective_as_list)

Chdf.head()
```

## Chapter section walkthrough

```python
Cwdf = pd.read_csv('../Data/Chapter section walkthrough.csv')
Cwdf.head()
```

```python
phrases_to_remove = [
    "How to Play Guitar",
    "How to Use Listen Mode",
    "How to Use Listening Mode",
    "How to Kill Shamblers",
    "If you're still having trouble with your picking, see the guide below! How to Play Guitar",
    "Trophy Guide | List of Trophies",
    "Crossbow: Location and Best Upgrades",
    "How to Kill The Rat King"
]

for phrase in phrases_to_remove:
    Cwdf['Mission Tips and Strategy'] = Cwdf['Mission Tips and Strategy'].str.replace(phrase, '', regex=False)

Cwdf['Mission Tips and Strategy'] = Cwdf['Mission Tips and Strategy'].str.replace(r'\*\*(.*?)\*\*', r'\1 - ', regex=True)


Cwdf.head()
```

```python
def download_image(image_url, filename):
    target_directory = '../Data/Images/chapter_section_images'
    local_path = os.path.join(target_directory, f"{filename}.jpg")

    try:
        response = requests.get(image_url, stream=True, timeout=10)
        response.raise_for_status()

        with open(local_path, 'wb') as out_file:
            for chunk in response.iter_content(chunk_size=8192):
                out_file.write(chunk)

        print(f"Downloaded: {local_path}")
        return local_path

    except requests.exceptions.RequestException as e:
        print(f"Error downloading {image_url}: {e}")
        return None
    except IOError as e:
        print(f"Error saving image {local_path}: {e}")
        return None
```

```python
for index, row in Cwdf.iterrows():
    image_url = row['Chapter Section Image']
    section_name = row['Chapter Section']
    download_image(image_url, section_name)
```

```python
for index, row in Cwdf.iterrows():
    image_url = row['Chapter Section Image']
    section_name = row['Chapter Section']
    local_path = os.path.join('../Data/Images/chapter_section_images', f"{section_name}.jpg")
    normalized_path = local_path.replace('\\', '/')
    Cwdf.loc[index, 'Chapter Section Image'] = normalized_path

Cwdf.head()
```

```python
Cwdf.to_csv('Chapter section walkthrough with mission tips and strategy and chapter section images.csv', index=False)
```

```python
def update_enemy_image_url(row):
    enemy_name = row['Enemy Name']
    image_path = f"../Data/Images/enemy_images/{enemy_name}.jpeg"
    return image_path

enemy_df['Enemy Image'] = enemy_df.apply(update_enemy_image_url, axis=1)

enemy_df.head()
```

```python
def extract_how_to_kill_text(guide_data_str):
    extracted_text = ""
    try:
        guide_data = ast.literal_eval(guide_data_str)

        if isinstance(guide_data, dict):
            for key, value in guide_data.items():
                if isinstance(value, list):
                    for item in value:
                        if isinstance(item, dict) and item.get('type') == 'paragraph' and 'content' in item:
                            extracted_text += item['content'] + " " 

    except (ValueError, SyntaxError) as e:
        print(f"Error evaluating guide data: {e}")
        return "" 

    return extracted_text.strip() 
```

```python
enemy_df['How To Kill Enemy'] = enemy_df['How To Kill Enemy'].apply(extract_how_to_kill_text)
enemy_df.head()
```

```python
enemy_df.to_csv('Enemy information.csv', index=False)
```

## Character EDA/Cleaning

```python
character_df = pd.read_csv('../Data/Characters.csv')
character_df.head()
```

```python
character_df['Character Main Image URL'] = character_df['Character Name'].apply(lambda x: f"Data/Images/character_images/{x} icon.jpeg")


character_df = character_df.rename(columns={
    'Character Main Image URL': 'Character Image',
    'Character Info': 'Character Information'
}).drop(columns=['Character Icon URL', 'Profile URL'])

character_df.head()
```

```python
character_info_df = pd.DataFrame({
    'Character name': character_df['Character Name'],
    'Character Image': character_df['Character Image']
})
character_info_df['Basic Information'] = ''
character_info_df['Profile and Backstory'] = ''
character_info_df['What happens to the character'] = ''
```

```python
character_info_df
```

```python
def populate_basic_information(character_name, character_df, character_info_df):
    character_data_row = character_df[character_df['Character Name'] == character_name]

    if not character_data_row.empty and 'Character Information' in character_data_row.columns:
        character_info_data = character_data_row['Character Information'].iloc[0]

    
        if isinstance(character_info_data, str):
            try:
                character_info_data = ast.literal_eval(character_info_data)
            except (ValueError, SyntaxError) as e:
                print(f"Could not eval character_info for {character_name}: {e}")
                character_info_data = {} 


        if isinstance(character_info_data, dict) and 'Basic Information' in character_info_data:
            basic_info_content = character_info_data['Basic Information']

 
            if isinstance(basic_info_content, dict) and 'tables' in basic_info_content:
                table_list = basic_info_content['tables']

         
                if isinstance(table_list, list) and table_list:
               
                     table_item = table_list[0]

               
                     if isinstance(table_item, dict) and 'table_data' in table_item:
                        table_data = table_item['table_data']
  
                        basic_info_str = ""
                        for row_data in table_data:
 
                             for key, value in row_data.items():

                                  basic_info_str += f"{value.strip()}; "


                        new_basic_info_keys = ['Age', 'Status', 'Occupation', 'Role', 'Voice Actor']
                        extracted_values = [v.strip() for v in basic_info_str.strip('; ').split(';')]

                        if len(extracted_values) == len(new_basic_info_keys):

                            formatted_basic_info_parts = []
                            for i in range(len(new_basic_info_keys)):
                                formatted_basic_info_parts.append(f"{new_basic_info_keys[i]}: {extracted_values[i]}")

                            formatted_basic_info = ", ".join(formatted_basic_info_parts)


                            char_info_index = character_info_df[character_info_df['Character name'] == character_name].index
                            if not char_info_index.empty:
                                character_info_df.loc[char_info_index[0], 'Basic Information'] = formatted_basic_info
                                print(f"Basic Information for {character_name} populated successfully.")
                            else:
                                print(f"{character_name}'s row not found in character_info_df DataFrame.")
                        else:
                            print(f"Warning: Number of extracted values ({len(extracted_values)}) does not match number of keys ({len(new_basic_info_keys)}) for {character_name}. Basic Information not reformatted with new keys.")
                            char_info_index = character_info_df[character_info_df['Character name'] == character_name].index
                            if not char_info_index.empty:
                                 character_info_df.loc[char_info_index[0], 'Basic Information'] = basic_info_str.strip('; ')
                                 print(f"Basic Information for {character_name} populated with raw extracted values.")
                            else:
                                 print(f"{character_name}'s row not found in character_info_df DataFrame.")


                     else:
                        print(f"Basic Information for {character_name}: Expected table_data key not found in table item.")

            elif isinstance(basic_info_content, dict) and 'paragraphs' in basic_info_content and basic_info_content['paragraphs']:
                 char_info_index = character_info_df[character_info_df['Character name'] == character_name].index
                 if not char_info_index.empty:
                     character_info_df.loc[char_info_index[0], 'Basic Information'] = basic_info_content['paragraphs']
                     print(f"Basic Information (paragraph data) for {character_name} populated successfully.")
                 else:
                     print(f"{character_name}'s row not found in character_info_df DataFrame.")
            else:
                print(f"Basic Information for {character_name} contains an unexpected structure (not tables or paragraphs).")
        else:
            print(f"Basic Information heading not found in {character_name}'s Character Info dictionary or dictionary is empty.")
    else:
        print(f"{character_name}'s row or 'Character Info' column not found in character_df DataFrame.")

```

```python
for character_name in character_info_df['Character name']:
    populate_basic_information(character_name, character_df, character_info_df)

character_info_df.head()
```

```python
def populate_what_happens(character_name, character_df, character_info_df):

  
    character_data_row = character_df[character_df['Character Name'] == character_name]

    
    if not character_data_row.empty and 'Character Information' in character_data_row.columns:
        character_info_data = character_data_row['Character Information'].iloc[0]

 
        if isinstance(character_info_data, str):
            try:
                character_info_data = ast.literal_eval(character_info_data)
            except (ValueError, SyntaxError) as e:
                print(f"Could not evaluate character_info string for {character_name}: {e}")
                character_info_data = {}

     
        if isinstance(character_info_data, dict):
            what_happens_heading = None
            for heading in character_info_data.keys():
                 if "What Happens to" in heading:
                     what_happens_heading = heading
                     break

          
            if what_happens_heading and what_happens_heading in character_info_data:
        
                 content = character_info_data[what_happens_heading]

                
                 if isinstance(content, dict) and 'paragraphs' in content and content['paragraphs']:
                      what_happens_text = content['paragraphs']

          
                      char_info_index = character_info_df[character_info_df['Character name'] == character_name].index
                      if not char_info_index.empty:
                          character_info_df.loc[char_info_index[0], 'What happens to the character'] = what_happens_text
                          print(f"What happens to {character_name} populated successfully.")
                      else:
                          print(f"{character_name}'s row not found in character_info_df DataFrame.")
                 else:
                    print(f"What happens to {character_name} contains an unexpected structure (not paragraphs).")
            else:
                print(f"What happens to heading not found in {character_name}'s Character Info dictionary.")
    else:
        print(f"{character_name}'s row or 'Character Info' column not found in character_df DataFrame.")
```

```python
for character_name in character_info_df['Character name']:
    populate_what_happens(character_name, character_df, character_info_df)

character_info_df.head()
```

## Safe EDA/Cleaning

```python
Safecodes_df = pd.read_csv('../Data/Safecode.csv')

Safecodes_df.head()
```

```python
Safecodes_df = Safecodes_df.drop(columns=['URL'])
Safecodes_df.head()
```

```python
safe_codes_df = pd.DataFrame({
    'Chapter': Safecodes_df['Chapter'],
    'Location': Safecodes_df['Location'],
    'Safe Code': Safecodes_df['Combination/Code']

})
safe_codes_df['Safe Image'] = ''
safe_codes_df['Finding the Safe Code'] = ''
```

```python
safe_codes_df.head()
```

```python
def update_safe_image_url(row):
    location_name = row['Location']
    image_path = f"../Data/Images/safe_images/{location_name}.jpg"
    return image_path

safe_codes_df['Safe Image'] = safe_codes_df.apply(update_safe_image_url, axis=1)

safe_codes_df.head()
```

```python
def process_safe_steps_and_images(row):
    steps_data_str = row['Steps to Safe']
    processed_text = ""

    try:
        steps_data = ast.literal_eval(steps_data_str)

        if isinstance(steps_data, list):
            all_paragraphs = []
            for step in steps_data:
                if isinstance(step, dict) and 'paragraphs' in step:
                    if isinstance(step['paragraphs'], list):
                        all_paragraphs.extend(step['paragraphs'])
                    elif isinstance(step['paragraphs'], str):
                        all_paragraphs.append(step['paragraphs'])

            processed_text = " ".join(all_paragraphs)

    except (ValueError, SyntaxError) as e:
        print(f"Error processing steps for {row['Location']}: {e}")
        processed_text = ""

    return processed_text

safe_codes_df['Finding the Safe Code'] = Safecodes_df.apply(process_safe_steps_and_images, axis=1)

safe_codes_df.head()
```

```python
locker_room_description = """In Chapter 2: Seattle Day 1 - The Tunnels, you will come across two rooms to the right of a long narrow pass in the tunnel. In one of those rooms, there will be a scrap of paper with a hint about where the code for the locked door to the locker room is.In the vending machine there is a red soda can that has a note taped to it that has the first 3 digits for the locker room lock. Despite being able to get these first three digits, there is no way to find the last two digits.
The Lock only has 5 buttons which must all be pressed, and no number can be repeated. Since you have the first 3 numbers of 1-5-2 there are only two other possible combinations to try.Once inside the locker room you will find scrap, ammo, and supplements."""

locker_room_index = safe_codes_df[safe_codes_df['Location'] == 'Locker Room'].index

if not locker_room_index.empty:
    safe_codes_df.loc[locker_room_index[0], 'Finding the Safe Code'] = locker_room_description.strip()
    print("Description for Locker Room updated successfully.")
else:
    print("Warning: Could not find 'Locker Room' in safe_codes_df to update.")
```

```python
def clean_finding_safe_code(text):
    phrases_to_remove = [
        "Click to see the Safe Combination ",
        "Weapons List | Where to Find All Weapons",
        "Pump Shotgun: Location and Best Upgrades"
    ]
    if isinstance(text, str):
        for phrase in phrases_to_remove:
            text = text.replace(phrase, '')
        # Replace multiple spaces with a single space
        text = re.sub(r'\s+', ' ', text).strip()
        return text
    return text

safe_codes_df['Finding the Safe Code'] = safe_codes_df['Finding the Safe Code'].apply(clean_finding_safe_code)

safe_codes_df.head()
```

```python
safe_codes_df.to_csv('Safecodes Information.csv', index=False)
```
