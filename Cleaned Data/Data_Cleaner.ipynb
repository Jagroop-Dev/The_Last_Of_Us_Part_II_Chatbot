```python
def format_main_objective_as_list(objectives_string):
    if isinstance(objectives_string, str) and objectives_string.strip():
        objectives_list = [obj.strip() for obj in objectives_string.split(',')]
        formatted_list = ""
        for i, objective in enumerate(objectives_list):
            if objective: 
                formatted_list += f"{i + 1}. {objective}\n"
        return formatted_list.strip() 
    else:
        return "" 
```

```python
phrases_to_remove = [
    "How to Play Guitar",
    "How to Use Listen Mode",
    "How to Use Listening Mode",
    "How to Kill Shamblers",
    "If you're still having trouble with your picking, see the guide below! How to Play Guitar",
    "Trophy Guide | List of Trophies",
    "Crossbow: Location and Best Upgrades",
    "How to Kill The Rat King"
]

for phrase in phrases_to_remove:
    Cwdf['Mission Tips and Strategy'] = Cwdf['Mission Tips and Strategy'].str.replace(phrase, '', regex=False)

Cwdf['Mission Tips and Strategy'] = Cwdf['Mission Tips and Strategy'].str.replace(r'\*\*(.*?)\*\*', r'\1 - ', regex=True)


Cwdf.head()
```

```python
def download_image(image_url, filename):
    target_directory = '../Data/Images/chapter_section_images'
    local_path = os.path.join(target_directory, f"{filename}.jpg")

    try:
        response = requests.get(image_url, stream=True, timeout=10)
        response.raise_for_status()

        with open(local_path, 'wb') as out_file:
            for chunk in response.iter_content(chunk_size=8192):
                out_file.write(chunk)

        print(f"Downloaded: {local_path}")
        return local_path

    except requests.exceptions.RequestException as e:
        print(f"Error downloading {image_url}: {e}")
        return None
    except IOError as e:
        print(f"Error saving image {local_path}: {e}")
        return None
```

```python
for index, row in Cwdf.iterrows():
    image_url = row['Chapter Section Image']
    section_name = row['Chapter Section']
    download_image(image_url, section_name)
```

```python
def update_weapon_image_url(row):
    weapon_name = row['Weapon Name']
    image_path = f"../Data/Images/weapon_images/{weapon_name}.jpg"
    return image_path

weapon_df['Weapon Image'] = weapon_df.apply(update_weapon_image_url, axis=1)
weapon_df.head()
```

```python
def format_best_upgrades(upgrades_data):
    if isinstance(upgrades_data, str):
        try:
            upgrades_list = ast.literal_eval(upgrades_data)
        except (ValueError, SyntaxError):
            if upgrades_data == "Name: N/A, Effect: N/A":
                 return "No upgrades for this weapon"
            return upgrades_data
    elif isinstance(upgrades_data, list):
        upgrades_list = upgrades_data
    else:
        return "" 

    if not isinstance(upgrades_list, list):
        if upgrades_list == "Name: N/A, Effect: N/A":
            return "No upgrades for this weapon"
        return "" 

    formatted_upgrades = []
    for upgrade in upgrades_list:
        if isinstance(upgrade, dict):
            name = upgrade.get('Name', 'N/A')
            effect = upgrade.get('Effect', 'N/A')
            formatted_upgrades.append(f"Name: {name}, Effect: {effect}")
    return ", ".join(formatted_upgrades)

weapon_df['Best Weapon Upgrades'] = weapon_df['Best Weapon Upgrades'].apply(format_best_upgrades)

weapon_df.head()
```

```python
def populate_basic_information(character_name, character_df, character_info_df):
    character_data_row = character_df[character_df['Character Name'] == character_name]

    if not character_data_row.empty and 'Character Info' in character_data_row.columns:
        character_info_data = character_data_row['Character Info'].iloc[0]

    
        if isinstance(character_info_data, str):
            try:
                character_info_data = ast.literal_eval(character_info_data)
            except (ValueError, SyntaxError) as e:
                print(f"Could not evaluate character_info string for {character_name}: {e}")
                character_info_data = {} 


        if isinstance(character_info_data, dict) and 'Basic Information' in character_info_data:
            basic_info_content = character_info_data['Basic Information']

 
            if isinstance(basic_info_content, dict) and 'tables' in basic_info_content:
                table_list = basic_info_content['tables']

         
                if isinstance(table_list, list) and table_list:
               
                     table_item = table_list[0]

               
                     if isinstance(table_item, dict) and 'table_data' in table_item:
                        table_data = table_item['table_data']
  
                        basic_info_str = ""
                        for row_data in table_data:
 
                             for key, value in row_data.items():

                                  basic_info_str += f"{value.strip()}; "


                        new_basic_info_keys = ['Age', 'Status', 'Occupation', 'Role', 'Voice Actor']
                        extracted_values = [v.strip() for v in basic_info_str.strip('; ').split(';')]

                        if len(extracted_values) == len(new_basic_info_keys):

                            formatted_basic_info_parts = []
                            for i in range(len(new_basic_info_keys)):
                                formatted_basic_info_parts.append(f"{new_basic_info_keys[i]}: {extracted_values[i]}")

                            formatted_basic_info = ", ".join(formatted_basic_info_parts)


                            char_info_index = character_info_df[character_info_df['Character name'] == character_name].index
                            if not char_info_index.empty:
                                character_info_df.loc[char_info_index[0], 'Basic Information'] = formatted_basic_info
                                print(f"Basic Information for {character_name} populated successfully.")
                            else:
                                print(f"{character_name}'s row not found in character_info_df DataFrame.")
                        else:
                            print(f"Warning: Number of extracted values ({len(extracted_values)}) does not match number of keys ({len(new_basic_info_keys)}) for {character_name}. Basic Information not reformatted with new keys.")
                            char_info_index = character_info_df[character_info_df['Character name'] == character_name].index
                            if not char_info_index.empty:
                                 character_info_df.loc[char_info_index[0], 'Basic Information'] = basic_info_str.strip('; ')
                                 print(f"Basic Information for {character_name} populated with raw extracted values.")
                            else:
                                 print(f"{character_name}'s row not found in character_info_df DataFrame.")


                     else:
                        print(f"Basic Information for {character_name}: Expected table_data key not found in table item.")

            elif isinstance(basic_info_content, dict) and 'paragraphs' in basic_info_content and basic_info_content['paragraphs']:
                 char_info_index = character_info_df[character_info_df['Character name'] == character_name].index
                 if not char_info_index.empty:
                     character_info_df.loc[char_info_index[0], 'Basic Information'] = basic_info_content['paragraphs']
                     print(f"Basic Information (paragraph data) for {character_name} populated successfully.")
                 else:
                     print(f"{character_name}'s row not found in character_info_df DataFrame.")
            else:
                print(f"Basic Information for {character_name} contains an unexpected structure (not tables or paragraphs).")
        else:
            print(f"Basic Information heading not found in {character_name}'s Character Info dictionary or dictionary is empty.")
    else:
        print(f"{character_name}'s row or 'Character Info' column not found in character_df DataFrame.")

```

```python
for character_name in character_info_df['Character name']:
    populate_basic_information(character_name, character_df, character_info_df)

character_info_df.head()
```

```python
def update_safe_image_url(row):
    location_name = row['Location']
    image_path = f"../Data/Images/safe_images/{location_name}.jpg"
    return image_path

safe_codes_df['Safe Image'] = safe_codes_df.apply(update_safe_image_url, axis=1)

display(safe_codes_df.head())
```

```python
def process_safe_steps_and_images(row):
    steps_data_str = row['Steps to Safe']
    processed_text = ""

    try:
        steps_data = ast.literal_eval(steps_data_str)

        if isinstance(steps_data, list):
            all_paragraphs = []
            for step in steps_data:
                if isinstance(step, dict) and 'paragraphs' in step:
                    if isinstance(step['paragraphs'], list):
                        all_paragraphs.extend(step['paragraphs'])
                    elif isinstance(step['paragraphs'], str):
                        all_paragraphs.append(step['paragraphs'])

            processed_text = " ".join(all_paragraphs)

    except (ValueError, SyntaxError) as e:
        print(f"Error processing steps for {row['Location']}: {e}")
        processed_text = ""

    return processed_text

safe_codes_df['Finding the Safe Code'] = Safecodes_df.apply(process_safe_steps_and_images, axis=1)

safe_codes_df.head()
```

```python
def clean_finding_safe_code(text):
    phrases_to_remove = [
        "Click to see the Safe Combination ",
        "Weapons List | Where to Find All Weapons",
        "Pump Shotgun: Location and Best Upgrades"
    ]
    if isinstance(text, str):
        for phrase in phrases_to_remove:
            text = text.replace(phrase, '')
        # Replace multiple spaces with a single space
        text = re.sub(r'\s+', ' ', text).strip()
        return text
    return text

safe_codes_df['Finding the Safe Code'] = safe_codes_df['Finding the Safe Code'].apply(clean_finding_safe_code)

safe_codes_df.head()
```
